```python
a = np.zeros((512, 512)).astype(np.uint8) #unsigned integer type needed by watershed
y, x = np.ogrid[0:512, 0:512]
a[((y-200)**2 + (x-100)**2 < 30**2)] = 1
a[((y-350)**2 + (x-400)**2 < 20**2)] = 2
a[((y-260)**2 + (x-200)**2 < 20**2)] = 3
plt.imshow(a, cmap='gray')# left plot in the image above
```

```python
b = np.zeros(
    (512, 512)).astype(np.uint8)  #unsigned integer type needed by watershed
y, x = np.ogrid[0:512, 0:512]
b[((y - 200)**2 + (x - 100)**2 < 30**2)] = 1
b[((y - 350)**2 + (x - 400)**2 < 25**2)
  & (np.arctan2(y - 350, x - 400) > 0)] = 1
b[((y - 350)**2 + (x - 400)**2 < 25**2)
  & (np.arctan2(y - 350, x - 400) < 0)] = 3
# b[((y - 350)**2 + (x - 400)**2 < 15**2)] = 3
b[((y - 260)**2 + (x - 200)**2 < 18**2)] = 2
plt.figure(figsize=(9, 9))
plt.imshow(b, cmap='gray')  # left plot in the image above
```
